```python
with open('../code/data/atkinpolys.json', 'r') as f:
    atkin_polys_raw = json.load(f)
atkin_polys = {int(p):atkin_polys_raw[p] for p in atkin_polys_raw}
```

```python
def eval_atk(x:int,l:int,p:int):
    if l not in atkin_polys:
        raise ValueError('l does not divide order of the Monster group')
    cs = atkin_polys[l]
    a,b1,b3= tuple([poly_eval_mod(ci[::-1],x,p) for ci in cs])
    b = (b1*pow(b3,3,p))%p
    return ((-a)%p,b)
```
