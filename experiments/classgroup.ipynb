```python
primes1 = primesBetween(5,2**10)
primes2 = primesBetween(2**10,2**14)
primesall = primes1+primes2
```

```python
def list_to_counts(l):
    counts = {l0:0 for l0 in l}
    for l0 in l:
        counts[l0]+=1
    return counts

def transpose_array(m):
    return [[r[i] for r in m] for i in range(len(m[0]))]

def supp(p):
    a = 0
    data = {}
    while a*a < 4*p:
        d = a*a-4*p
        d0,c = discfac(d)
        data[a] = (d0,c,d)
        a+=1
    return data

def unlabelled_data(p):
    return transpose_array([list(t) for t in supp(p).values()])
```

```python
ssprimes = primesBetween(2,36)+[41]+[47+12*i for i in range(3)]
ssds = {}
ssds_list_wrep = []
for p in ssprimes:
    sp = supp(p)
    dsp = []
    for a in sp:
        d0,c,d = sp[a]
        dsp+=[d0*(c0**2) for c0 in divisors(c)]
    dsp = list(set(dsp))
    dsp.sort(key=abs)
    ssds[p]=dsp
    ssds_list_wrep+=dsp
ssds_counts = list_to_counts(ssds_list_wrep)
ssds_counts_inv = {c:[] for c in ssds_counts.values()}
for d in ssds_counts:
    ssds_counts_inv[ssds_counts[d]].append(d)
```

```python
ds = [int(d) for d in df.d.unique()]
ds.sort(key = abs)
d2dc = {d:discfac(d) for d in ds}
df['dfun']=df.d.apply(lambda x:d2dc[int(x)][0])
df['cond']=df.d.apply(lambda x:d2dc[int(x)][1])
conds = {int(c):primefact(int(c)) for c in df.cond.unique()}
```

```python
def j_int_c_simp(d):
    heegner_ds = [-3,-4,-7,-8,-11,-19,-43,-67,-163]
    d0, c = discfac(d)
    if d0 not in heegner_ds:
        return False
    if c <4:
        return True
    if d0 < -7:
        pfc = primefact(c)
        return len(pfc)==1 and max(pfc.values())==1
    elif d0 >-3:
        c = c//gcd(c,2)
        return len(divisors(c))<3
    else:
        if c % 3 == 0:
            return len(divisors(c//3))<3
        elif c % 2 == 0:
            return len(divisors(c//2))<3
        else:
            return len(divisors(c))<3

    

def hpfails(d):
    return not j_int_c_simp(d)

    

```
